```python
import booknlp_utils
import pdnc_utils
import gen_utils
```

```python
for novel in novels:
    print(novel)
    ent_df = booknlp_utils.read_entdf(novel)
#     ent_df['prop'].value_counts()

    qdf = booknlp_utils.read_qdf(novel)
    qents = qdf['char_id'].unique().tolist()
    qentdf = ent_df[ent_df['COREF'].isin(qents)]


#     len(qentdf['COREF'].unique())

#     Counter(qentdf['prop'])

    coref2names = {}
    for _, row in ent_df.iterrows():
        coref = row['COREF']
        prop = row['prop']
        text = row['text']
        if coref not in coref2names:
            coref2names[coref] = {}
        if prop not in coref2names[coref]:
            coref2names[coref][prop] = []

        coref2names[coref][prop].append(text)

    num_prop = 0
    num_pron = 0
    num_nom = 0
    coref2name = {}
    for e in qentdf['COREF'].unique():
        if 'PROP' in coref2names[e]:
            coref2name[e] = ['PROP', Counter(coref2names[e]['PROP']).most_common()]
            num_prop += 1
        elif 'NOM' in coref2names[e]:
            coref2name[e] = ['NOM', Counter(coref2names[e]['NOM']).most_common()]
            num_nom += 1
        else:
            coref2name[e] = ['PRON', Counter(coref2names[e]['PRON']).most_common()]
            num_pron += 1


    print("Speaker types (PROP, PRON, NOM): ", num_prop, num_pron, num_nom)

#     coref2name

    coref_matches = pd.read_csv('../booknlpen/pdnc_output/'+novel+'/coref_matches.csv')

#     coref_matches

    ent_matches = pd.read_csv('../booknlpen/pdnc_output/'+novel+'/booknlp_matched_ents.csv')

#     ent_matches

#     len(ent_df['COREF'].unique())

#     Counter(ent_matches['cat'])

#     Counter(ent_matches['pdncID'])

#     len(ent_matches['COREF'].unique())

#     coref_matches[coref_matches['pdncID']!=-1]

    prop_corefs = set()
    for c, e in coref2name.items():
        if e[0]=='PROP':
            prop_corefs.add(c)

#     len(prop_corefs)

    res_corefs = set(coref_matches[coref_matches['pdncID']!=-1]['COREF'])

#     len(res_corefs)

    for ue in prop_corefs-res_corefs:
        print(ue, coref2name[ue])
    
    print("--"*10)
#     coref2name[0]

#     cinfo = pdnc_utils.read_char_info(novel)

#     cinfo['id2names']
```
